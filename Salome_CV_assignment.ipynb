# **1. Fintech: How Computer Vision Can Improve Payments in Nigeria**

Nigeria's payment system is moving quickly toward cashless options. Today, most people have mobile phones and can make instant transfers. That makes Nigeria a good place to use computer vision (CV)—technology that lets cameras and software recognize faces, codes, objects, and actions—to make payments both easy and safe.

##**1.1 Biometrics (Face and Fingerprint via Camera)**

What it is: Using your face, fingerprint, or palm captured by a camera to prove it's really you.

**Why it helps:**

**Less fraud:** A face or fingerprint is hard to steal compared to a PIN. If a bank app asks for a quick face match before a big transfer, it lowers the chance of someone else sending your money.

**Faster logins and approvals:** Instead of typing passwords, a quick face check can approve a payment in seconds.

**Works with what Nigeria already has:** Banks already use the Bank Verification Number (BVN) to identify customers. Face checks can add an extra layer of protection during sign-up and high-risk payments, while still being convenient.

**Important note:** Nigeria's Data Protection Act (2023) treats biometrics as sensitive data. Any app using face or fingerprint matching must protect people's data, ask for consent, and be transparent about how images are stored and used (Nigeria Data Protection Act, 2023).

## **1.2 QR Codes (Scan to Pay)**

#### **What it is:** The merchant shows a black-and-white square (QR code). You open your phone camera, scan it, and confirm payment in your bank or wallet app.

#### **Why it helps:**

**Simple and cheap:** A shop only needs a printed code or a simple screen—no expensive card machine.

**Works in markets and small shops:** Even micro-businesses can accept digital payments.

**Backed by national standards:** The Central Bank of Nigeria (CBN) created rules so all banks and wallets can use the same QR system, and NIBSS launched NQR, a national “scan-to-pay” platform. This means a code from one bank should work with many apps (CBN, 2021; NIBSS, n.d., 2021).

####**Where computer vision helps:**

The phone's camera and software find and read the QR code even in poor lighting or if the paper is bent.

Extra checks can spot tampered or fake codes (for example, if someone sticks a fake code on top of a real one), reducing fraud.

##**1.3 “Swipe and Pay” (Fast, Low-Touch Checkout)**

What it is: Moving quickly through checkout with very little typing or touching—like tapping a card/phone, or using a simple hand gesture to confirm.

#### **Why it helps:**

**Speed:** Vision systems can help self-checkout lines move faster by recognizing items or confirming a customer's intent with a quick, clear gesture (like holding up a palm to confirm).

**Less contact:** Useful in crowded places or public transport, where people want to move fast with minimal touch.

###**Putting it all together:**
Computer vision can add convenience (quick scans, quick approvals) and security (face checks, tamper detection) on top of Nigeria's existing payment rails and QR standards.

# **2. Health Tech: Using Computer Vision to Expand Access to Quality Care**
Nigeria's health-tech startups are building the digital “pipes” of care—electronic records, logistics for blood and oxygen, and telemedicine. Computer vision can ride on these rails to reach more people with better, faster service.

### **2.1 Screening and Triage at Scale (Finding Who Needs Help First)**
- **Chest X-rays for TB:** The **World Health Organization (WHO)** now recommends **computer-aided detection (CAD)** software to help read chest X-rays for **tuberculosis (TB)** screening. In places with few radiologists, CAD can quickly flag likely TB cases so people get confirmatory tests sooner (WHO, 2021).  
- **Diabetic eye disease:** Simple smartphone cameras plus AI can help **screen for diabetic retinopathy**. Primary-care clinics can capture eye images, use software to grade risk, and escalate urgent cases to specialists. This reduces blindness by catching problems early.  
- **Skin and wound images:** Patients can upload photos of **rashes, infections, or wounds**. CV tools can help triage—“urgent,” “see a clinician soon,” or “monitor.” Doctors still make the final decision, but the first sort happens quickly.  

### **2.2 Telemedicine that Works on Everyday Phones**
- **Low-cost imaging:** Many clinics already use smartphones for photos and basic imaging. CV can check if an image is **clear and usable** (e.g., “retake—too blurry”), saving time for both patients and clinicians.  
- **Digitizing paperwork:** CV can read text from paper lab results or prescriptions (like a smart scanner), storing the data directly in the clinic's **electronic medical record (EMR)**. Platforms such as **Helium Health** already provide EMRs and hospital systems that can plug in these features (Helium Health, n.d., 2019/2025).  
- **Better logistics:** Health logistics platforms like **LifeBank** move blood, oxygen, and other essentials to hospitals. CV can help **verify labels, check seals, and track inventory** during transport, improving safety and reducing errors (LifeBank, n.d.).  

### **2.3 Safety, Privacy, and Trust**
Health images and biometrics are **very sensitive**. Nigeria’s **Data Protection Act (2023)** requires companies to:  
- collect only what is necessary,  
- protect it with strong security,  
- explain how it is used, and  
- give people rights over their data.  

Startups should also keep a **human in the loop** (clinicians reviewing AI outputs) and clearly tell users what the AI can and cannot do.

Computer vision can make **payments** in Nigeria **safer, faster, and more inclusive** through biometrics, QR code “scan-to-pay,” and low-touch checkout. # **3. Restoring a Corrupted Group Image**

### Using Computer Vision image processing, restore the image.

*  Write a code to **show the pictures original color ** and **correct the contrast and brightness**
-	What contrast number (alpha) and brightness number (beta) restored the picture to its original state.

```python
# Importing Libraries
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Loading image
img = cv2.imread("/content/Corrupted Student Picture.png")#Read the image
plt.imshow(img) # show the image - The image is displayed here as BGR (Blue,Green,Redb)
```

```python
# Displaying the actual form of the image - Which is the RGB channels
img_rgb = cv2_imshow(img) # RGB (Red,Green,Blue)
```

```python
# Adjusting the Brightness and Contrast of the image
alpha = 0.9 # Contrast control
beta = 0.8 # Brightness control
adjusted_img = cv2.convertScaleAbs(img, alpha = alpha, beta = beta)
plt.imshow(adjusted_img)
```

##	***What contrast number (alpha) and brightness number (beta) restored the picture to its original state.***

- Contrast -  (alpha): 0.9

- Brightness - (beta): 0.8
